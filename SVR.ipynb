```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
```

```python
data=pd.read_csv(r"E:\3rdyear\2nd_sem\3200\datafiles\cg.csv")
```

```python
data.head(3)
```

```python
x =data[["CGPA"]]
y=data["Salary_Package_LPA"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
from sklearn.svm import SVR
sr=SVR(kernel="poly",C=0.1)
sr.fit(x_train,y_train)
```

```python
sr.score(x_test,y_test)*100,sr.score(x_train,y_train)
#overfitting
```

```python
sns.scatterplot(x="CGPA",y="Salary_Package_LPA",data=data)
plt.plot(data["CGPA"],sr.predict(x),color="red")
plt.show()
```

```python
sr1=SVR(kernel="linear")
sr1.fit(x_train,y_train)
```

```python
sr1.score(x_test,y_test)*100
```

```python
sns.scatterplot(x="CGPA",y="Salary_Package_LPA",data=data)
plt.plot(data["CGPA"],sr1.predict(x),color="red")
plt.show()
```
