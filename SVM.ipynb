```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv(r"E:\3rdyear\2nd_sem\3200\datafiles\placement_dataset.csv")
```

```python
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
x=dataset[["cgpa","score"]]
y=dataset["placed"]
```

```python
sns.scatterplot(x="cgpa",y="score",data=dataset,hue="placed")
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
sr = SVC(kernel="linear")
sr.fit(x_train,y_train)
```

```python
sr.score(x_test,y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sr)
plt.show()
```
