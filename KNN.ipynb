```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r"E:\3rdyear\2nd_sem\3200\datafiles\classification_dataset.csv")
dataset.head(3)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="age",y="income",hue="purchased",data=dataset)
plt.show()
```

```python
x= dataset[["age","income"]]
y=dataset["purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn= KNeighborsClassifier(n_neighbors=15,p=1)
kn.fit(x_train,y_train)
```

```python
kn.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(max_depth=4)
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)
```
